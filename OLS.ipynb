### IV.4. For each portfolio of N = 11, 12, 13, ..., 50 securities obtained by adding one by one each security to the initial portfolio, calculate: 
### ‚Ä¢ its Œ≤p
### ‚Ä¢ its non-diversifiable risk and its diversifiable risk

```python
rest = list(range(1, 51))

for i in tirages:
    rest.remove(i)
print(rest)

tirages_new_tmp = list(tirages)

betas_ptf = []
non_div_risk = []
div_risk = []

for i in rest:
    
    tirages_new_tmp.append(i)
    tirages_new = np.array(tirages_new_tmp)
    print(tirages_new)
    
    rdt_ptf = df_log_returns.iloc[:,tirages_new]
    rdt_ptf.head()
    
    wgt_new = np.ones(len(tirages_new)) * 1/len(tirages_new)
    rdt_ptf_equi = rdt_ptf * wgt_new
    rdt_ptf_equi_sum = rdt_ptf_equi.sum(axis=1)
    rdt_ptf_equi_sum
    
    # betas of different portfolios

    company_data_equi_bis = np.array(rdt_ptf_equi_sum)

    benchmark_data_equi_bis = np.array(benchmark_data)

    company_data_equi_bis = sm.add_constant(company_data_equi_bis) 

    regall = sm.OLS(benchmark_data_tmp,company_data_equi_bis).fit()

    betas_ptf.append(regall.params[1])

    
    lst = []
    
    for j in range(rdt_ptf_equi.shape[1]):

        company_data_equi_bis = np.array(rdt_ptf_equi.iloc[:,j])

        benchmark_data_equi_bis = np.array(benchmark_data)

        company_data_equi_bis = sm.add_constant(company_data_equi_bis) 

        regall = sm.OLS(benchmark_data_tmp,company_data_equi_bis).fit()

        lst.append(regall.params[1])


    betas_np_bis = np.array(lst)
    
    
    
    lst_var = []

    for j in range(rdt_ptf_equi.shape[1]):

        expected_rdt = betas_np_bis[j] * df_log_returns.iloc[:,0]

        real_rdt = rdt_ptf.iloc[:,j]  

        lst_var.append(np.var(expected_rdt - real_rdt))

    omega = np.diag(lst_var)
    
    
    div = wgt_new.T @ omega @ wgt_new
    div_risk.append(div)
    
    non_div_part_1 = ((betas_np_bis.T @ wgt_new ) ** 2)
    non_div_part_2 = np.var(df_log_returns.iloc[:,0])
    non_div = non_div_part_1 * non_div_part_2
    non_div_risk.append(non_div)
 
  

    print("the non diversifiable risk is", non_div)
    
    
    print("the diversifiable risk is", div)
```
